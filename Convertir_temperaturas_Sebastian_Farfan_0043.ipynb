## 1. Fórmulas de Conversión

Para movernos entre estas dos escalas, necesitamos aplicar una relación lineal. Aquí tienes las expresiones matemáticas:

* **De Grados Celsius (°C) a Grados Fahrenheit (°F):**
    $$F = (C \times \frac{9}{5}) + 32$$

* **De Grados Fahrenheit (°F) a Grados Celsius (°C):**
    $$C = (F - 32) \times \frac{5}{9}$$

## 2. Ejemplos Resueltos Paso a Paso

### Ejemplo A: Convertir $30°C$ a $°F$ (Un día caluroso)
1.  **Identificar la fórmula:** Usamos $F = (C \times \frac{9}{5}) + 32$.
2.  **Sustituir el valor:** $F = (30 \times \frac{9}{5}) + 32$.
3.  **Multiplicar por la fracción:** $30 \times 1.8 = 54$.
4.  **Sumar la constante:** $54 + 32 = 86$.
* **Resultado:** $30°C$ equivalen a **$86°F$**.

### Ejemplo B: Convertir $50°F$ a $°C$ (Un día fresco)
1.  **Identificar la fórmula:** Usamos $C = (F - 32) \times \frac{5}{9}$.
2.  **Sustituir el valor:** $C = (50 - 32) \times \frac{5}{9}$.
3.  **Restar la constante:** $50 - 32 = 18$.
4.  **Multiplicar por la fracción:** $18 \times (\frac{5}{9}) \rightarrow \frac{90}{9} = 10$.
* **Resultado:** $50°F$ equivalen a **$10°C$**.

## 3. Código en Python

Este programa sigue una estructura modular, separando la lógica de los cálculos de la interacción con el usuario.

```python
def convertirCF():
    """Convierte de Celsius a Fahrenheit"""
    celsius = float(input("\nIngrese los grados Celsius a convertir: "))
    fahrenheit = (celsius * 9/5) + 32
    print(f"--- {celsius}°C equivalen a {fahrenheit:.2f}°F ---")

def convertirFC():
    """Convierte de Fahrenheit a Celsius"""
    fahrenheit = float(input("\nIngrese los grados Fahrenheit a convertir: "))
    celsius = (fahrenheit - 32) * 5/9
    print(f"--- {fahrenheit}°F equivalen a {celsius:.2f}°C ---")

def menu():
    """Muestra el menú de opciones"""
    print("\n--- CONVERSOR DE TEMPERATURA ---")
    print("1. Convertir grados Centígrados a Fahrenheit")
    print("2. Convertir grados Fahrenheit a Centígrados")
    print("3. Salir")
    opcion = input("Seleccione una opción (1-3): ")
    return opcion

# Cuerpo principal del programa
while True:
    seleccion = menu()

    if seleccion == "1":
        convertirCF()
    elif seleccion == "2":
        convertirFC()
    elif seleccion == "3":
        print("Saliendo del programa... ¡Hasta luego!")
        break
    else:
        print("Opción no válida. Intente de nuevo.")
```

### Explicación del código:
* **`while True`**: Crea un bucle infinito que permite realizar múltiples conversiones sin tener que reiniciar el programa.
* **`float()`**: Usamos esto en el `input` porque las temperaturas suelen llevar decimales.
* **`break`**: Es la instrucción que rompe el bucle cuando el usuario elige la opción 3.
* **`f"{valor:.2f}"`**: Se utiliza para redondear el resultado a 2 decimales y que se vea más limpio en pantalla.
